Replace and add Cape-Cod area population survey data for a total of 6 species

```python
dfCC6 = pd.read_pickle('../PklJar/dfCensus_Cape_Cod_6spp.pkl')
```

```python
dfCC6.head(20)
```

```python
df_MAG = dfFeatures[['FRDH', 'FRDV', 'OTTH', 'OTTV']]
```

```python
df_SUN = dfFeatures[['Kp_index', 'Sunspot_No', 'Dst_index', 'AE_index', 'Proton_flux_gt30MeV', 'Proton_flux_gt60MeV']]
```

```python
commonprefs = ['Pred_', 'ampDiff_', 'vel_', 'move_']
suffxs = ['min', 'max', 'std', 'med']
tide_cols = ['%s%s' %(prefx, sufx) for prefx in commonprefs for sufx in suffxs]
```

```python
df_TIDE = dfFeatures[tide_cols]
```

```python
df_WIND = dfFeatures[['EasterlyWS', 'SoutherlyWS']]
```

```python
df_WATER = dfFeatures[['WTMP', 'WVHT', 'APD', 'DPD']]
```

```python
dfLabels.head()
```

```python
df_TARGETS = dfLabels.iloc[:, 1:7]
```

```python
df_TOTAL = df_TARGETS.sum(axis=1)
```

```python
df_TOTAL
```

```python
CapeCodDataDict = {'Features': {'CENSUS': dfCC6,
                               'MAG': df_MAG,
                               'SUN': df_SUN,
                               'TIDE': df_TIDE,
                               'WIND': df_WIND,
                               'WATER': df_WATER},
                  'Targets': {'By_SPP': df_TARGETS,
                             'TOTAL': df_TOTAL}
                  }
```

```python
pd.testing.assert_index_equal(dfCC6.index, df_MAG.index)
```

```python
pd.testing.assert_index_equal(df_MAG.index, df_SUN.index)
```

```python
idxref = df_TOTAL.index
for key, val in CapeCodDataDict['Features'].items():
    print('testing ', key)
    idx = val.index
    pd.testing.assert_index_equal(idxref, idx)
```

```python
pd.testing.assert_index_equal(df_TOTAL.index, df_TARGETS.index)
```

```python
with open('../PklJar/CapeCodDataDict', 'wb') as f:
    pickle.dump(CapeCodDataDict, f, protocol=pickle.HIGHEST_PROTOCOL)
```
